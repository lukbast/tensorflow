```python
## Insepect the data
import os

# Recursive function to print folder structure
def print_dirs(path:str,n_indents=0):
    indent="    "
    for dir_name in os.listdir(path):
        if os.path.isdir(os.path.join(path, dir_name)):
            print(f"{indent*n_indents}{dir_name}")
            print_dirs(os.path.join(path, dir_name), n_indents+1)
```

## Preprocessing the images

```python
# Turn images into a tensors and normalize them
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator


# set the seed
tf.random.set_seed(42)

# Preprocess data

# set up image data generators
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Import data from directories and turn it into batches
train_batch = train_datagen.flow_from_directory(directory=train_path,
                                                  batch_size=32,
                                                  target_size=(224,224),
                                                  class_mode="binary",
                                                  seed=42 )

test_batch = test_datagen.flow_from_directory(directory=test_path,
                                                  batch_size=32,
                                                  target_size=(224,224),
                                                  class_mode="binary",
                                                  seed=42 )
```

```python
# check if GPU is ready
tf.config.list_physical_devices("GPU")
```

```python
# set the seed
tf.random.set_seed(42)
# Build a CNN model
model_1 = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(filters=10,
                          kernel_size=3,
                          activation="relu",
                          input_shape=(224,224,3)),
    tf.keras.layers.Conv2D(10,3, activation="relu"),
    tf.keras.layers.MaxPool2D(pool_size=2,
                             padding="valid"),
    tf.keras.layers.Conv2D(10,3, activation="relu"),
    tf.keras.layers.Conv2D(20,3, activation="relu"),
    tf.keras.layers.MaxPool2D(2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

# Compile our CNN
model_1.compile(loss="binary_crossentropy",
               optimizer="adam",
               metrics=["accuracy"])

# Fit the model
history_1 = model_1.fit(train_batch,
                        epochs=5,
                        steps_per_epoch=len(train_batch),
                        validation_data=test_batch,
                        validation_steps=len(test_batch))
```

```python
model_1.evaluate(test_batch);
```

```python
model_1.summary()
```

Look like we have 10 classes in this data set. Therefore this time we will deal with multiclass classification problem.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_path = "./data/04_Computer_Vision_and_CNNs/10_food_classes_all_data/train"
test_path = "./data/04_Computer_Vision_and_CNNs/10_food_classes_all_data/test"

# Set up train and test data generators
train_gen = ImageDataGenerator(rescale=(1./255))
test_gen = ImageDataGenerator(rescale=(1./255))

# Preprocess the images
train_batch = train_gen.flow_from_directory(directory=train_path,
                                            target_size=(224,224), 
                                            seed=42)
test_batch = test_gen.flow_from_directory(directory=test_path,
                                           target_size=(224,224),
                                           seed=42)
```

### Create baseline model

```python
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Activation
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam

# Create checkpoints function
def create_checkpoint_callback(models_name):
    model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
        filepath=f"./models/04_Computer_Vision_and_CNNs/{models_name}",
        save_weights_only=False,
        monitor='val_accuracy',
        mode='max',
        save_best_only=True)
    return model_checkpoint_callback
```

```python
# set random seed
tf.random.set_seed(42)

# Set models name
model_name = "model_1_multiclass.h5"

# Create a model
model_1 = Sequential([
    Conv2D(10, 3, input_shape=(224,224,3), activation="relu"),
    MaxPool2D(),
    Conv2D(10, 3, activation="relu"),
    MaxPool2D(),
    Conv2D(10, 3, activation="relu"),
    MaxPool2D(),
    Flatten(),
    Dense(10, activation="softmax")
])

# Compile the model
model_1.compile(loss="categorical_crossentropy",
                optimizer="adam",
                metrics=["accuracy"]
               )

# fit the model
model_1.fit(train_batch,
            steps_per_epoch=len(train_batch),
            epochs=10,
            validation_data=test_batch,
            validation_steps=len(test_batch),
            callbacks=[create_checkpoint_callback(model_name)]
           )
```

Looks like model is overfitting. Remove one layer.

```python
# set random seed
tf.random.set_seed(42)

# Set models name
model_name = "model_2_multiclass.h5"

# Create a model
model_2 = Sequential([
    Conv2D(10, 3, input_shape=(224,224,3), activation="relu"),
    MaxPool2D(),
    Conv2D(10, 3, activation="relu"),
    MaxPool2D(),
    Flatten(),
    Dense(10, activation="softmax")
])

# Compile the model
model_2.compile(loss="categorical_crossentropy",
                optimizer="adam",
                metrics=["accuracy"]
               )

# fit the model
model_2.fit(train_batch,
            steps_per_epoch=len(train_batch),
            epochs=10,
            validation_data=test_batch,
            validation_steps=len(test_batch),
            callbacks=[create_checkpoint_callback(model_name)]
           )
```
