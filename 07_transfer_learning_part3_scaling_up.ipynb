### Check data

```python
utils.print_dir("./data/07_transfer_learning_part3_scaling_up/101_food_classes_10_percent/")
```

```python
train_dir = "./data/07_transfer_learning_part3_scaling_up/101_food_classes_10_percent/train"
test_dir = "./data/07_transfer_learning_part3_scaling_up/101_food_classes_10_percent/test"


len(utils.os.listdir(train_dir)), len(utils.os.listdir(test_dir))
```

This time we have 101 classes.

```python
train_data, test_data = utils.create_data_loaders(train_dir, test_dir)
```

### Setup model

```python
# Create augmentation layer
aug_layer = utils.create_data_aug_layer()
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential, applications, Model, layers
from tensorflow.keras.layers.experimental import preprocessing

tf.random.set_seed(42)
# Setup base model
base_model = applications.EfficientNetB0(include_top=False)
base_model.trainable = False
#Setup model architecture with trainable top layers
inputs = layers.Input(shape=(224,224, 3), name="input_layer")
x = aug_layer(inputs),
x = base_model(x, training=False)
x = layers.GlobalAvgPool2D(name="glob_avg_pool_2d")(x)
outputs = layers.Dense(101, activation="softmax", name="output_layer")(x)

model_0 = Model(inputs, outputs)
```

```python
# compile model
model_0.compile(loss="categorical_crossentropy",
               optimizer="adam",
               metrics=["accuracy"])
```

```python
tf.random.set_seed(42)
# Callbacks
callbacks = [
    utils.create_checkpoint_callback("checkpoints/07_transfer_learning_part3_scaling_up/model_0/checkpoint.ckpt",
                                     best_only=True,
                                     weights_only=True),
    utils.create_tensorboard_callback("logs/07_transfer_learning_part3_scaling_up/",
                                     "EffNet_baseline")
]
# Fit the model
history_0 = model_0.fit(train_data,
           steps_per_epoch=len(train_data),
           epochs=5,
           validation_data=test_data,
           validation_steps=int(0.15 * len(test_data)),
           callbacks=callbacks)
```

```python
# Evaluate on the whole data set
model_0.evaluate(test_data) 
```

```python
# lets prepare exising model for fine tuning 
utils.create_model_for_fine_tuning(base_model, 5)

# recompile model
model_0.compile(loss="categorical_crossentropy",
               optimizer="adam",
               metrics=["accuracy"])

callbacks = [
    utils.create_checkpoint_callback("checkpoints/07_transfer_learning_part3_scaling_up/model_1/checkpoint.ckpt",
                                     best_only=True,
                                     weights_only=True),
    utils.create_tensorboard_callback("logs/07_transfer_learning_part3_scaling_up/",
                                     "EffNet_fine_tune_5layers")]

# fine-tune model
history_1 = model_0.fit(train_data,
           steps_per_epoch=len(train_data),
           epochs=10,
           validation_data=test_data,
           validation_steps=int(0.15 * len(test_data)),
           callbacks=callbacks,
          initial_epoch=5)
```
